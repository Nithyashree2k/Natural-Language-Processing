```python
##tensorflow >2.0
# Apply One Hot Encoding
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### corpus
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
### Vocabulary size
voc_size=500
```

#### One Hot Representation

```python
# Print indexes with 1 in the vector of each document
onehot_repr=[one_hot(words,voc_size)for words in sent]
print(onehot_repr)
```

### Word Embedding Representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

Apply Padding Techniques to fix the length of each document

```python
## pre padding
sent_length=8  # Fix max length to 8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## post padding
sent_length=8  # Fix max length to 8
embedded_docs1=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs1)
```

```python
## 10 feature dimesnions (convert each word/value to vector representation)
dim=10
```

```python
#  CREATE MODEL
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length)) # Embedding layer - (vocab_size,no_of_features,total_sentence_length)
model.compile('adam','mse')  # adam optimizer, mean squared error
```

```python
model.summary()
```

```python
# each point in above document is represented as a vector
model.predict(embedded_docs[0])
```
